```python
def get_list_id():
    list_id = []
    for i in range(0, len(train_set)):
        list_id.append(train_set[i].id)
    for i in range(0, len(dev_set)):
        list_id.append(dev_set[i].id)
    for i in range(0, len(test_set)):
        list_id.append(test_set[i].id)
    return list_id
```

```python
# non if non medata
# read the metadata if metadata and add to the dataset
def read_metadata(path):

    json_files = [pos_json for pos_json in os.listdir(path) if pos_json.endswith('.json')]

    metadata = {}

    for i in range(len(json_files)):
        f = open(path+"/"+json_files[i], encoding='utf-8')
        # print(json_files[i])
        data = json.load(f,strict=False)
        # if ( json_files[i] == "123534.json"):
        #     print(type(data["director"]))
        #     print("yes")
        if "director" in data:
            # print(data)
            if type(data["director"]) == list:
                metadata.update({json_files[i][:-5]: [d['name'] for d in data["director"]]})
            else:
                metadata.update({json_files[i][:-5]: data["director"]["name"]})
        elif "creator" in data:
            if type(data["creator"]) == list:
                metadata.update({json_files[i][:-5]: [d['name'] for d in data["creator"]]})
            else:
                metadata.update({json_files[i][:-5]: data["creator"]["name"]})
        else:
            metadata.update({json_files[i][:-5]: "none"})
        
        f.close()
        # metadata.update({json_files[i][:-5]: data["director"]["name"]})

    return metadata
```
